```python
import cv2
```

```python
image_path = '/content/image_test1.png'
def preprocess_image(image_path):
  image = cv2.imread(image_path, cv2.CALIB_CB_NORMALIZE_IMAGE)
  #apply thresholding to improve readabiloity
  _, processed_image = cv2.threshold(image, 150,255, cv2.THRESH_BINARY_INV)
  return processed_image
```
